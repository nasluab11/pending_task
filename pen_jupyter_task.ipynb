```python
import pandas as pd

# Create a Pandas Series with the top 5 programming languages and their popularity index
programming_languages = pd.Series({
    'Python': 100,
    'JavaScript': 95,
    'Java': 90,
    'C++': 85,
    'C#': 80
})

# Extract the third item from the series (index starts from 0, so index 2 represents the third item)
third_item = programming_languages.iloc[2]
print(third_item)
```

```python
#exercise 2
# Create a Pandas Series with the given data
series_data = pd.Series([1, 2, 3, 4, 5])

# Calculate the sum, mean, and standard deviation of the series
sum_series = series_data.sum()
mean_series = series_data.mean()
std_dev_series = series_data.std()

print("Sum:", sum_series)
print("Mean:", mean_series)
print("Standard Deviation:", std_dev_series)
```

# Numpy universal functions


```python
# Create an array of angles
angles = np.array([0, 30, 45, 60, 90, 180])

# Conversion of degrees into radians using deg2rad function
radians = np.deg2rad(angles)

# Sine of angles
print('Sine of angles in the array:')
sine_value = np.sin(radians)
print(np.sin(radians))

# Inverse Sine of sine values (converting back to degrees)
print('Inverse Sine of sine values:')
degrees_sine = np.rad2deg(np.arcsin(sine_value))
print(degrees_sine)

# Sine hyperbolic of angles
print('Sine hyperbolic of angles in the array:')
sineh_value = np.sinh(radians)
print(np.sinh(radians))

# Inverse Sine hyperbolic (converting back to original values)
print('Inverse Sine hyperbolic:')
inverse_sineh = np.sin(sineh_value)
print(inverse_sineh)

# Hypot function demonstration (calculating hypotenuse of a right triangle)
base = 4
height = 3
hypotenuse = np.hypot(base, height)
print('Hypotenuse of right triangle is:', hypotenuse)
```

# Statistical:

```python
# Construct a weight array
weight = np.array([50.7, 52.5, 50, 58, 55.63, 73.25, 49.5, 45])

# Minimum and maximum
print('Minimum and maximum weight of the students:')
print(np.amin(weight), np.amax(weight))

# Range of weight (i.e., max weight - min weight)
print('Range of the weight of the students:')
print(np.ptp(weight))

# Percentile
print('Weight below which 70% of students fall:')
print(np.percentile(weight, 70))

# Mean
print('Mean weight of the students:')
print(np.mean(weight))

# Median
print('Median weight of the students:')
print(np.median(weight))

# Standard deviation
print('Standard deviation of weight of the students:')
print(np.std(weight))

# Variance
print('Variance of weight of the students:')
print(np.var(weight))

# Average (Mean)
print('Average weight of the students:')
print(np.average(weight))


```
